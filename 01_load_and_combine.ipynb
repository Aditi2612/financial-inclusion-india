```python
all_data = []

for f in files:
    temp = pd.read_excel(f, header=5)
    temp = temp.dropna(axis=1, how='all')      # drop fully-empty columns
    all_data.append(temp)
    print(f"Loaded {f[-20:]} — {temp.shape}")

df = pd.concat(all_data, ignore_index=True)
print("COMBINED:", df.shape)
```

```python
df.to_csv("bsr1_combined.csv", index=False)
print("saved")
```
